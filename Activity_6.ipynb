```python
import pandas as pd
data = pd.read_csv(r'/content/Netflix Userbase.csv')
df = pd.DataFrame(data)
print(df)

```

```python
from numpy import mean
pd.pivot_table( df,
               index = 'Country',
                values = 'Monthly Revenue',
              columns = ['Gender','Age'],
              aggfunc = mean,
                )
```

```python
df.dropna(subset = favs,
          inplace = True
          )
df
```

```python
X = pd.DataFrame(df['cintura'])
y = pd.DataFrame(df['peso'])

lm = linear_model.LinearRegression()

model_lm = lm.fit(X, y)
model_lm
```

```python
print('The intercept is \t :\t',         model_lm.intercept_)
print('The coefficients are \t : \t',    model_lm.coef_)
print('Model score (R-squared) is : \t', model_lm.score(X, y))
```

```python
modelA  = sm.OLS.from_formula("peso ~ cintura",
                              data=df
                             )
resultA = modelA.fit()
resultA.summary()
```

```python
predict_values = resultA.fittedvalues.to_frame()
RESI_A = resultA.resid

study = [ df['peso'],
         df['cintura'],
         predict_values,
         RESI_A
         ]

study  #is a list
```

```python
study_table = pd.concat(study,
                        axis=1
                       )
study_table
```

```python
study_table.columns =['','peso',
                      'cintura',
                      'PredictValues',
                      'RESI_Age',''
                      ] #renaming the columns

study_table['difference'] = study_table['peso'] - study_table['PredictValues']

study_table
```

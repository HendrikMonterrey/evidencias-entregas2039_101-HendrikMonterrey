```python
import pandas as pd
import numpy as np

info = {'Company': ['Ford', 'Ford', 'VW', 'BMW', 'Cooper', 'Cooper'],
         'Stars': [1, 2, np.nan, 2, 1, 1],
         'Weight': ['China', 'Mexico', 'Mexico', None, 'China', pd.NaT],
         'Length': [40, 50, 30, np.nan, 45, None]
         }
```

```python
df = pd.DataFrame(data = info)
df.head()
```

```python
df.isnull().any()
```

```python
sm= df.Stars.mean()
```

```python
df['Stars'].fillna(value = sm, inplace = True)
df
```

```python
ov = df.Weight.mode()[1]
ov
```

```python
df['Weight'].fillna(value = ov, inplace = True)
df
```

```python
df['Length'].fillna(value = df.Length.mean(), inplace = True)
df
```

```python
average_monthly_revenue = df['Monthly Revenue'].mean()
```

```python
count_monthly_revenue = df['Monthly Revenue'].count()
```

```python
total_revenue = average_monthly_revenue * count_monthly_revenue
```

```python
print(f"Total Revenue: {total_revenue}")
```

```python
country_revenue = df.groupby('Country')['Monthly Revenue'].sum()
country_share = (country_revenue / total_revenue) * 100
```

```python
plt.figure(figsize=(8, 8))
plt.pie(country_share, labels=country_share.index, autopct='%1.1f%%', startangle=140)
plt.title('Country Share of Total Revenue')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.boxplot([df[df['Device'] == device]['Age'] for device in df['Device'].unique()],
            labels=df['Device'].unique())
plt.title('Box Plot of Age by Device')
plt.xlabel('Device')
plt.ylabel('Age')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
